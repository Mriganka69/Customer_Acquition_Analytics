```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv("updated_customer_data_with_cac.csv")

# Set a clean visual style
sns.set(style="whitegrid")
```

```python
# Choose a channel to filter by (change this manually)
selected_channel = 'All'  # Change to 'Email Marketing', 'Social Media', etc.

# Apply filter
if selected_channel == 'All':
    filtered_df = df.copy()
else:
    filtered_df = df[df['Marketing_Channel'] == selected_channel]
```

## Display Summary Metrics

```python
total_spend = filtered_df['Marketing_Spend'].sum()
total_customers = filtered_df['New_Customers'].sum()
average_cac = filtered_df['CAC'].mean()

print(f"ðŸ“Š Dashboard Summary for: {selected_channel}")
print(f"----------------------------------------")
print(f"ðŸ’° Total Marketing Spend: ${total_spend:,.2f}")
print(f"ðŸ‘¥ Total New Customers: {total_customers}")
print(f"ðŸ“‰ Average CAC: ${average_cac:.2f}")
```

## Bar Chart: Average CAC by Channel

```python
channel_cac = df.groupby("Marketing_Channel")['CAC'].mean().sort_values()

plt.figure(figsize=(8, 4))
sns.barplot(x=channel_cac.index, y=channel_cac.values, palette="Blues_d")
plt.title("Average CAC by Marketing Channel")
plt.ylabel("CAC")
plt.xticks(rotation=30)
plt.show()
```

## Scatter Plot: Spend vs New Customers

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=filtered_df, x='Marketing_Spend', y='New_Customers', hue='Marketing_Channel', palette='Set1')
plt.title("Marketing Spend vs New Customers")
plt.xlabel("Marketing Spend")
plt.ylabel("New Customers")
plt.legend(title="Channel")
plt.show()
```
