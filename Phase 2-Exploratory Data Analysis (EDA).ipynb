```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("customer_acquisition_cost_dataset.csv")

# Set a consistent style
sns.set(style="whitegrid")

# 1. Distribution of Marketing Spend
plt.figure(figsize=(10, 4))
sns.histplot(df['Marketing_Spend'], kde=True, bins=30, color='skyblue')
plt.title("Distribution of Marketing Spend")
plt.xlabel("Marketing Spend")
plt.ylabel("Frequency")
plt.show()
```

```python
# 2. Distribution of New Customers
plt.figure(figsize=(10, 4))
sns.histplot(df['New_Customers'], kde=True, bins=30, color='salmon')
plt.title("Distribution of New Customers")
plt.xlabel("New Customers")
plt.ylabel("Frequency")
plt.show()
```

```python
# 3. Count of Entries per Marketing Channel
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='Marketing_Channel', palette='Set2')
plt.title("Count of Entries by Marketing Channel")
plt.xticks(rotation=30)
plt.ylabel("Count")
plt.show()
```

```python
# 4. Average Spend and Customers by Channel
channel_group = df.groupby('Marketing_Channel').agg({
    'Marketing_Spend': 'mean',
    'New_Customers': 'mean'
}).reset_index()
# Bar plot: average marketing spend per channel
plt.figure(figsize=(8, 4))
sns.barplot(data=channel_group, x='Marketing_Channel', y='Marketing_Spend', palette='Blues_d')
plt.title("Average Marketing Spend by Channel")
plt.xticks(rotation=30)
plt.ylabel("Avg Spend")
plt.show()
```

```python
# Bar plot: average new customers per channel
plt.figure(figsize=(8, 4))
sns.barplot(data=channel_group, x='Marketing_Channel', y='New_Customers', palette='Greens_d')
plt.title("Average New Customers by Channel")
plt.xticks(rotation=30)
plt.ylabel("Avg New Customers")
plt.show()
```

```python
# 5. Scatter Plot: Marketing Spend vs New Customers
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Marketing_Spend', y='New_Customers', hue='Marketing_Channel', palette='Set1')
plt.title("Marketing Spend vs New Customers")
plt.xlabel("Marketing Spend")
plt.ylabel("New Customers")
plt.legend(title='Channel')
plt.show()
```

```python
# 6. Correlation Heatmap (for numeric columns)
plt.figure(figsize=(6, 4))
sns.heatmap(df[['Marketing_Spend', 'New_Customers']].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
