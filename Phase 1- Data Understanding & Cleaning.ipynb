# Phase 1: Data Understanding & Cleaning

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("customer_acquisition_cost_dataset.csv")

# Display the first few rows of the dataset
print("First 5 rows:")
print(df.head())
```

```python
# Basic info about the dataset
print("\nDataset Info:")
print(df.info())

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Check for duplicate rows
print("\nDuplicate Rows:")
print(df.duplicated().sum())

# Summary statistics
print("\nSummary Statistics:")
print(df.describe())

# Check unique values in categorical columns
print("\nUnique Marketing Channels:")
print(df['Marketing_Channel'].value_counts())
```

```python
# Check for outliers using boxplots
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x='Marketing_Spend', data=df)
plt.title('Boxplot of Marketing Spend')

plt.subplot(1, 2, 2)
sns.boxplot(x='New_Customers', data=df)
plt.title('Boxplot of New Customers')

plt.tight_layout()
plt.show()
```
